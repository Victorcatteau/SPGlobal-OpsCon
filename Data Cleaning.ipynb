# DATA CLEANING

## Import Data

```python
import pandas as pd
#Don't forget to change the path
df = pd.read_csv(r'C:\Users\Alu\Desktop\Criteo\userjourney_samplefile_8757 (1).tsv2', delimiter='\t')

# I copy the data to keep the initial data intact
df_copy = df.copy()
```

```python
df_copy
```

## Remove Listing event

```python
df_copy = df_copy[df_copy['a.event_name']!='Listing'].reset_index(drop=True)
```

```python
df_copy
```

## Remove Category event

```python
df_copy[df_copy['a.event_name']!='Category'].reset_index(drop=True)
```

```python
df_copy
```

## Remove Sales Ducplicate 

```python
list_index = df_copy[(df_copy['a.event_name'] == 'Sales') & (df_copy['a.item_id'].isna()) & (df_copy['a.arbitrage_id'].isna())].index.values.astype(int)
df_copy=df_copy.drop(list_index).reset_index(drop=True)
```

```python
df_copy
```

## Remove Customer with a number of session higher than gamma

```python
gamma = 50 
df_copy = df_copy.groupby(['a.user_id']).filter(lambda x: x['a.user_session_id'].nunique() < gamma).reset_index(drop=True)
```

```python
df_copy
```

## Remove useless Display event

A useless display is a display that is not linked to a sales or a click by a arbitrage_id
This operation can takes several minutes

```python
list_index=df_copy[df_copy['a.arbitrage_id'].notna()].groupby(['a.arbitrage_id']).filter(lambda x: x['a.event_name'].nunique() == 1).index.values.astype(int)
df_copy= df_copy.drop(list_index).reset_index(drop=True)
```

```python
df_copy
```

## Remove usless ItemPageView event

A useless ItemPageView is a ImagePageView that is not linked to a Sales or a basket by a item_id
Or a ImagePageView that have no item_id

```python
list_index_1 = df_copy[(df_copy['a.event_name'] == 'ItemPageView') & (df_copy['a.item_id'].isna())].index.values.astype(int)
list_index_2 =df_copy[df_copy['a.arbitrage_id'].isna()].groupby(['a.item_id']).filter(lambda x: x['a.event_name'].nunique() == 1).index.values.astype(int)
list_index = list(set(list_index_1) | set(list_index_2))
df_copy= df_copy.drop(list_index).reset_index(drop=True)
```
